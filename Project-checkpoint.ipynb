```python
import joblib
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
features = ETC.drop('Prevalence', axis=1)
labels = ETC['Prevalence']
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
rf = RandomForestRegressor()
parameters = {
    'n_estimators': [5, 50, 250,400],
    'max_depth': [2,4, 8, 16, 32, None]
}

cv = GridSearchCV(rf, parameters, cv=5)
cv.fit(features, labels.values.ravel())

print_results(cv)
```

```python
from sklearn.model_selection import train_test_split

X, y = features,labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # Random state keeps the split consistent
print(X_train.shape, X_test.shape)
```

```python
model = RandomForestRegressor(n_estimators = 400, max_depth = 4) # Create the model
model.fit(X_train, y_train) # Train it (this syntax looks the same for all sklearn models)
model.score(X_test, y_test) # Show a score
```

```python
RF = model.predict(X_test)

plt.scatter(y_test, RF, alpha=0.3)
plt.xlabel('y_test')
plt.ylabel('prediction')
plt.title('Random  Forest Prediction plot')
```
