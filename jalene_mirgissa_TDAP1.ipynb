```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.linspace(0,1,100)
#s= -p * np.log(p)
plt.plot(p,-p * np.log(p))
plt.xlabel ("p")
plt.ylabel ("s")
plt.show()
```

```python
T = np.linspace(0,100,200)
#s= -p * np.log(p)
plt.plot(T, np.exp(-1/T))
plt.xlabel ("T")
plt.ylabel ("g")
plt.show()
```

```python
E = np.linspace(-1,30,1000)
#s= -p * np.log(p)
l= (np.exp(3*(E+1)))-1
m = (np.exp (30*(E+1)))-1
b3 = 1/l
b30 = 1/m
plt.plot(E,b3,E,b30)
plt.plot(E[:5:],b3[:5:],"*")
plt.plot(E[::10],b30[::10],'+')
plt.xlabel ("E")
plt.ylabel ("b3, b30")
plt.show()
```
